```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import plotly.graph_objects as go
```

```python
#Trabajar con otro tipo de datos Fifa
fifa = pd.read_csv("/content/drive/MyDrive/Simulacion/fifa_datos.csv")
#imprimir los primeros 5 datos del archivo
fifa.head(5)
```

```python
hola=fifa["Weight"].str.split("lbs",expand=True)
fifa["Weight"] = pd.to_numeric(hola[0])
print(type(fifa["Weight"][0]))
```

```python
# Generar un grafico de cual es su pie diestro

p1 = fifa.loc[fifa["Weight"] < 125 ].count()[0]
p2 = fifa.loc[(fifa["Weight"] >= 125)&(fifa["Weight"] < 150 )].count()[0]
p3 = fifa.loc[(fifa["Weight"] >= 150) & (fifa["Weight"] < 175)].count()[0]
p4 = fifa.loc[fifa["Weight"] > 175 ].count()[0]
print(p1)
plt.figure()

etiquetas = ['<125', '>125&<150','>150&<175','>175']
colores = ['#F63715', '#aabbcc',"#EFEB16","#83F615"]
plt.pie([p1,p2,p3,p4], labels=etiquetas, colors=colores, autopct='%.2f %%')
plt.title('Clasificacion Weight por Libras')
plt.show()
```

```python
o1 = fifa.loc[fifa["Overall"] == 40 ].count()[0]
o2 = fifa.loc[fifa["Overall"] == 50 ].count()[0]
o3 = fifa.loc[fifa["Overall"] == 60 ].count()[0]
o4 = fifa.loc[fifa["Overall"] == 70 ].count()[0]
o5 = fifa.loc[fifa["Overall"] == 80 ].count()[0]
o6 = fifa.loc[fifa["Overall"] == 90 ].count()[0]
```

```python
etiquetas = ['40', '50', '60','70', '80','90']
valores =(o1,o2,o3,o4,o5,o6)
plt.figure()
barras = plt.bar(etiquetas, valores)
plt.title('Clasificacion Overall')

plt.show()
```
